```python
import pandas as pd
df1 = pd.read_csv('StreamingHistory0.csv')
df2 = pd.read_csv('StreamingHistory1.csv')

df = pd.concat([df1, df2])
df.head()
```

```python
#Divido la columna endTime en columnas relevantes, tambien pasando de milisegundos de la columna msPlayed a segundos y minutos
df['year'] = df['endTime'].str.split('-').str[0]
df['month'] = df['endTime'].str.split('-').str[1]
df['seconds_played'] = df['msPlayed']/1000
df['minutes_played'] = df['seconds_played']/60

df.drop('msPlayed', inplace=True, axis=1)
df.head()
```

```python
# Paso las horas a tiempo Bonaerense
df['endTime'] = pd.to_datetime(df['endTime']).dt.tz_localize('UTC')
df['endTime'] = df['endTime'].dt.tz_convert('America/Argentina/Buenos_Aires')
df.head()
```

```python
df.info()
```

```python
#Convierto los tipos de datos
df['artistName'] = df['artistName'].astype(str)
df['trackName'] = df['trackName'].astype(str)
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df.info()
```

```python
#Filtro las canciones que se escucharon por menos de medio minuto ya que no son representativos
indexes = df[df['minutes_played'] < 0.5].index
df.drop(index=indexes, inplace=True, axis=0)
df.describe()
```

```python
#Export a un nuevo dataset para mandar a Tableau
df.to_csv('spotify_data_4tableau.csv')
```
